```python
import numpy as np
n = 20
x = np.zeros(n+1)
k = 0
x[k] = 2
while k < n:
  x[k+1] = 1 + np.sin(x[k])
  k += 1
print('x = ',x)
print('cos(x_0) = ',np.cos(x[0]))
print('cos(x_1) = ',np.cos(x[1]))
L = np.max(np.abs(np.cos(x[0:1])))
print('L = ',L)
tol = 1e-4
k = np.log( (1-L)*tol/np.abs(np.cos(x[0])-np.cos(x[1])) )/np.log(L)
print('k = ',k)
k = int(k)+1
print('k = ',k)
print(x[k]-x[n])
```
